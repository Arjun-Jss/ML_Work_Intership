**METHOD 1**

```python
#count vectorizer - feature extraction 
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer(stop_words = 'english')
x_train_vect = vect.fit_transform(x_train)
x_test_vect = vect.transform(x_test)
```

```python
from sklearn.svm import SVC #support vector machine 
#SVC - support vector classifier 
#SVC - similar to logistic regrssion 
#SVC gives better accuracy 
```

```python
model = SVC()
model.fit(x_train_vect,y_train)
```

```python
y_pred  = model.predict(x_test_vect)
```

```python
from sklearn.metrics import accuracy_score 
```

```python
accuracy_score(y_pred,y_test)
```

**METHOD 2** 

```python
#using pipeline ( pipelineis used to combine two ML algorithms / two different processes together)
```

```python
#pipeline - SVC + count vectorizer 
from sklearn.pipeline import make_pipeline
text_model = make_pipeline(CountVectorizer(),SVC())

text_model.fit(x_train,y_train)
y_pred = text_model.predict(x_test)
y_pred
```

**USING JOBLIB TO  SAVE THE MODE**L
